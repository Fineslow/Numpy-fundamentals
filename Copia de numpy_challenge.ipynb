Create a filter array with values < 20.



```
import numpy as np
ageInformation = np.array([18,20,19,23,19,20])
for element in ageInformation:
  if element <= 20:
    np.append(ageInformation, True)
  else:
    np.append(ageInformation, False)
print(ageInformation)
```
